# Data Collection

**What this notebook does:**  
Loads the Customer Support Tickets dataset (downloaded from Kaggle) saved at `data/raw/customer_support/`.  
This dataset is the foundation for our **task type classifier** and **priority predictor**.

**Dataset:** Customer Support Ticket Dataset  
**Source:** https://www.kaggle.com/datasets/suraj520/customer-support-ticket-dataset  
**Local path:** `../data/raw/customer_support/customer_support_tickets.csv`

```python
import pandas as pd

DATA_PATH = "../data/raw/customer_support/customer_support_tickets.csv"

df = pd.read_csv(DATA_PATH)

print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")
print()
print("--- First 10 Rows ---")
print(df.head(10).to_string())
print()
print("--- Column Names ---")
for col in df.columns:
    print(f"  {col}")
```

```python
print("=== Data Types ===")
print(df.dtypes)
print()
print("=== Null Count Per Column ===")
print(df.isnull().sum())
print()
print("=== Unique Value Count Per Column ===")
print(df.nunique())
```

```python
print("=== Ticket Type Distribution ===")
print(df["Ticket Type"].value_counts())
print()
print("=== Ticket Priority Distribution ===")
print(df["Ticket Priority"].value_counts())
```

```python
import os

path = "../data/raw/customer_support/customer_support_tickets.csv"
file_exists  = os.path.exists(path)
file_size_mb = os.path.getsize(path) / (1024 * 1024)

print(f"File exists : {file_exists}")
print(f"File size   : {file_size_mb:.2f} MB")
print()
print(f"Dataset ready. Shape: {df.shape}. File size: {file_size_mb:.2f} MB")
```
